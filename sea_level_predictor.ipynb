```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
# Read data

df = pd.read_csv('epa-sea-level.csv')
df
```

```python
fig, ax = plt.subplots(figsize=(19,8))
ax.set_xlabel('Year')
ax.set_ylabel('Sea Level (inches)')
ax.set_title('Rise in Sea Level')

plt.scatter(df['Year'],df['CSIRO Adjusted Sea Level'], label='data')

#First line of best fit
res = linregress(df['Year'],df['CSIRO Adjusted Sea Level'])
Xv = np.arange(df['Year'].min(),2050,1)
Yv = Xv * res.slope + res.intercept
plt.plot(Xv, Yv, 'r', label='fitted line')


#Second line of best fit
df_2000 = df[df['Year'] >= 2000]
res2 = linregress(df_2000['Year'],df_2000['CSIRO Adjusted Sea Level'])
Xv2 = np.arange(2000,2050,1)
Yv2 = Xv2 * res2.slope + res2.intercept
plt.plot(Xv2, Yv2, 'g', label='fitted line 2')

plt.legend()

fig.savefig('Rise in Sea Level.png')
```
